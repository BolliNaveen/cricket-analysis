****T20 WORLDCUP Cricket Data Preprocessing****

```python
#importing libraries
import pandas as pd
import json
```

**1.Processing Match Results**

```python
#opening json file
with open("Music/t20_json_files/t20_wc_match_results.json") as f:
    data=json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
#renaming scorecard to match id
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
#connecting matches with match ids 

match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

**2.Processing Batting summary**

```python
#opening the batting summary json file
with open('Music/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])
  
df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
#creating a new column out or not based on the dismissal column
df_batting['out/not_out'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
df_batting.head(11)
```

```python
#removing the dismissal column
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

```python
#removing special characters
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
#interconnecting the both the dataframes using match id
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.shape
```

**3. Processing Bowling Summary**

```python
#reading the bowling summary
with open('Music/t20_json_files/t20_wc_bowling_summary.json') as f:
    data=json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
#converting the json format files into the dataframe
df_bowling=pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
#creating a match id column to the bowling data and interconnecting the data tables
df_bowling['match_id']=df_bowling['match'].map(match_ids_dict)
df_bowling.head()
```

**4. Processing Players Information**

```python
#loading the players json data into dataframe
with open('Music/t20_json_files/t20_wc_player_info.json') as f:
    data=json.load(f)
df_players=pd.DataFrame(data)
print(df_players.shape)
df_players.head(10)
    
```

```python
#cleaning the special characters from the players data
df_players['name']=df_players['name'].apply(lambda x: x.replace('â€',''))
df_players['name']=df_players['name'].apply(lambda x: x.replace('†',''))
df_players['name']=df_players['name'].apply(lambda x: x.replace('\xa0',''))
df_players.head(10)
```
